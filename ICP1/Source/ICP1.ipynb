```python
import seaborn as sns
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
path='/content/drive/MyDrive/Data.csv'
df= pd.read_csv(path)
```

```python
df.tail(10)
```

```python
uniqueValues = df['diagnosis'].unique()
print('Unique elements in the column "diagnosis" ')
print(uniqueValues)
```

```python
df.groupby(['diagnosis']).sum().plot(kind='pie', y='id', autopct='%1.1f%%', title='Breast cancer diagnosis')
```

```python
correlations = df.corr()
sns.heatmap(correlations)
sns.set(rc = {'figure.figsize':(15,10)})
plt.show()
```

```python
memory=df.memory_usage(deep=True)
print('Memory usage by each column')
print('____________________________________')
print(memory)
print('************************************')
sum=df.memory_usage(deep=True).sum()
print('Total memory used')
print(sum)
```
